```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/language.csv')
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

## CountVectorizer, it does the main work of converts a collection of text documents into a matrix of token counts (bag-of-words model).
## It transforms text into numerical feature vectors that can be used in machine learning models.
```

```python
# create a CountVectorizer object
vectorizer = CountVectorizer()

#sample text data
data2= ["love data science", "love machine learning "]

# fit the vectorizer and transform the data
vectorized_data = vectorizer.fit_transform(data2)

# get the vocabulary (unique words)
print(vectorizer.get_feature_names_out())

# print the vectorized data
print(vectorized_data.toarray())

# NOTE : BAG OF WORDS ONLY FOCUSES ON WORDS NOT ITS SEQUENCE AND ODERING
```

NOW WE HAVE TO CREATE `BAG OF WORDS FOR ENTIRE DATASET` .

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['language'].value_counts()
```

```python
x=np.array(data['Text'])
y=np.array(data['language'])
```

```python
print (x), print(y)
```

now text wala column (x ki bat ho rhi h ) ka BAG OF WORDS banega. ARRAY => NUMBERS

```python
cv = CountVectorizer()
X = cv.fit_transform(x)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
print(X_train)
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
user =input ( "Enter a Text: " )
data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
